```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
import matplotlib.pyplot as plt

# Calcular el índice de calidad del clúster para diferentes valores de MinPts
def graph_minPts(eps):
    min_pts_values = range(1, 11)
    silhouette_scores = []

    for min_pts in min_pts_values:
        dbscan = DBSCAN(eps=0.5, min_samples=min_pts)
        labels = dbscan.fit_predict(df)

        # Verificar si hay al menos dos grupos
        unique_labels = len(set(labels))
        if unique_labels > 1:
        silhouette_scores.append(metrics.silhouette_score(df, labels))
    else:
        silhouette_scores.append(None)
```

```python
import plotly.graph_objects as go
# Graficar el resultado
# Crear el gráfico interactivo en Plotly
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(min_pts_values), y=silhouette_scores, mode='lines+markers'))

fig.update_layout(
    title='Método del Codo para DBSCAN',
    xaxis=dict(title='MinPts'),
    yaxis=dict(title='Silhouette Score'),
    showlegend=False
)

fig.show()
```

```python
def plot_knee_curve(df, min_sampl):
    # Fit Nearest Neighbors model
    nbrs = NearestNeighbors(n_neighbors=min_sampl).fit(df)
    
    # Find the k-neighbors of a point
    neigh_dist, neigh_ind = nbrs.kneighbors(df)
    
    # Sort the neighbor distances in ascending order
    sort_neigh_dist = np.sort(neigh_dist, axis=0)
    
    # Extract the distances to the 5th nearest neighbor
    k_dist = sort_neigh_dist[:, min_sampl - 1]
        
    # Use KneeLocator to find the knee point
    kneedle = KneeLocator(x=range(1, len(neigh_dist) + 1), y=k_dist, S=1.0,
                          curve="concave", direction="increasing", online=True)
    
    fig = kneedle.plot_knee()
    
    # Return the figure without showing it
    return fig

min_sampl = 1
fig = plot_knee_curve(df, min_sampl)
```

```python
x= 'age'
y = 'experience'
def graph_bubble(df, x, y, labels):
    # Get labeled data
    df['labels'] = labels
    
    # Convert 'labels' column to categorical
    df['labels'] = pd.Categorical(df['labels'])
    
    # Count of values based on x and y values
    bubble_sizes = df.groupby([x, y, 'labels']).size().reset_index(name='count')
    
    # Join values of bubble_sizes
    df = pd.merge(df, bubble_sizes, on=[x, y, 'labels'], how='left')
    
    # Create bubble chart in plotly
    fig = px.scatter(df, x=x, y=y, 
                     size='count', color='labels',
                     color_discrete_sequence=px.colors.qualitative.Set1,
                     title='Bubble Chart con Clusters',
                     labels={'x': 'Variable X', 'y': 'Variable Y'})
    
    return fig


graph_bubble(df, x, y, labels)
```

```python
# Visual Python: ANOVA
from scipy import stats
def vp_confidence_interval(var, confidence_level=0.95):
    try: sr = pd.Series(var)
    except: return np.nan
    return stats.t.interval(confidence_level, df=sr.count()-1, loc=sr.mean(), scale=sr.std() / np.sqrt(sr.count()) )
def vp_sem(var):
    try: sr = pd.Series(var)
    except: return np.nan
    return sr.std() / np.sqrt(sr.count())
```
